```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            print("Antrian kosong")
            return
        print("Memanggil nomor:", self.front.data)
        self.front = self.front.next
        if self.front is None:
            self.rear = None

    def peek(self):
       if self.front:
            print("Antrian sekarang:", self.front.data)
       else:
            print("Kosong")

    def display(self):
        temp = self.front
        if not temp:
            print("Antrian kosong")
            return
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

q = Queue()
nomor = 1

while True:
    print("\n=== SISTEM ANTRIAN ===")
    print("1. Ambil Nomor Antrian")
    print("2. Panggil Antrian")
    print("3. Lihat Antrian Depan")
    print("4. Tampilkan Semua")
    print("5. Keluar")

    pilih = int(input("Pilih: "))


    if pilih == 1:
        print("Nomor antrian:", nomor)
        q.enqueue(nomor)
        nomor += 1
    elif pilih == 2:
        q.dequeue()
    elif pilih == 3:
        q.peek()
    elif pilih == 4:
        q.display()
    elif pilih == 5:
        break
```
